# LinearRegression
### 공부 시간에 따른 시험 점수

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('LinearRegressionData.csv')
```

```python
dataset.head() #head: 데이터의 상위 5개를 불러옴
```

```python
X = dataset.iloc[:,:-1].values #처음부터 마지막 컬럼 직전까지의 데이터(독립 변수(피처))
y = dataset.iloc[:,-1].values #마지막 컬럼 데이터(종속데이터(레이블))
```

```python
X,y
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression() #객체 생성
reg.fit(X, y) #학습(모델 생성)
```

```python
y_pred = reg.predict(X) #X에 대한 예측값
y_pred
```

```python
plt.scatter(X,y,color = 'blue') #산점도
plt.plot(X,y_pred, color = 'green') #선 그래프
plt.xlabel('hours')
plt.ylabel('score')
plt.show()
```

```python
print('9시간 공부했을 떄 예상 점수:', reg.predict([[9]]))
```

```python
reg.coef_ #기울기
```

```python
reg.intercept_ #y절편
```

### 데이터 세트 분리

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('LinearRegressionData.csv')
```

```python
dataset
```

```python
X = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.2, random_state=0)
```

```python
X, len(X)
```

```python
X_train, len(X_train)
```

```python
X_test, len(X_test)
```

### 분리된 데이터를 통한 모델링

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
```

```python
reg.fit(X_train, y_train)
```

### 데이터 시각화(훈련세트)

```python
plt.scatter(X_train,y_train,color = 'blue') #산점도
plt.plot(X_train,reg.predict(X_train), color = 'green') #선 그래프
plt.xlabel('hours')
plt.ylabel('score')
plt.show()
```

### 데이터 시각화(테스트 세트)

max_iter = 훈련 세트 반복 횟수(Epoch) 수

eta0 = 학습률(learn rate)

```python
plt.scatter(X_test,y_test,color = 'blue') #산점도
plt.plot(X_train,reg.predict(X_train), color = 'green') #선 그래프
plt.xlabel('hours')
plt.ylabel('score')
plt.show()
```

```python
reg.coef_
```

```python
reg.intercept_
```

### 모델 평가

```python
reg.score(X_test, y_test) #테스트 세트를 통한 모델 평가(0 ~ 1)
```

```python
reg.score(X_train, y_train)
```

## 경사 하강법(Gradient Descent)

```python
from sklearn.linear_model import SGDRegressor #SGD: Stochastic Gradient Descent(확률적 경사 하강법)
sr = SGDRegressor(max_iter = 1000, eta0 = 0.001, random_state=0,verbose=1)
sr.fit(X_train, y_train)
```

```python
plt.scatter(X_train,y_train,color = 'blue') #산점도
plt.plot(X_train,sr.predict(X_train), color = 'green') #선 그래프
plt.xlabel('hours')
plt.ylabel('score')
plt.show()
```

```python
sr.coef_
```

```python
sr.intercept_
```

```python
sr.score(X_train, y_train)
```

```python
sr.score(X_test,y_test)
```
